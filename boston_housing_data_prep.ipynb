```python
# remove outlier
q1 = df['MEDV'].quantile(0.25)
q3 = df['MEDV'].quantile(0.75)
iqr = q3 - q1
low = q1 - 1.5*iqr
high = q3 + 1.5*iqr

filter = (df['MEDV'] >= low) & (df['MEDV'] <= high)
clean_df = df.loc[filter]
clean_df
```
